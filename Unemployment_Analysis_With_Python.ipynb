# Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px #for data visualization
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")
```

# Data Preprocessing

```python
data.columns = ["States","Date","Frequency","Estimated Unemployement Rate","Estimated Unemployed","Estimated Labour Participation Rate","Region","Longitude","Latitude"]
```

```python
data
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(8,6))
sns.heatmap(data.corr())
plt.show()
```

```python
plt.title("Unemployement Rate In India")
sns.histplot(x="Estimated Unemployed",data=data)
plt.show()
```

```python
# searching for estimated number of employees according to different areas

data.columns = ["States","Date","Frequency","Estimated Unemployement Rate"
               ,"Estimated Unemployed","Estimated Labour Participation Rate",
              "Region","Longitude","Latitude"]

plt.title("Unemployement Rate In India")
sns.histplot(x="Estimated Unemployed",hue="Region",data=data)
plt.show()
```

Creating a dashboard for the estimated unemployement rate

```python
unemployement = data[["States","Region","Estimated Unemployement Rate"]]
figure = px.sunburst(unemployement,path=["States","Region"],
                     values="Estimated Unemployement Rate",
                     width=700,height=700,color_continuous_scale="RdY1Gn",
                     title="Unemployment Rate In India")
figure.show()
```
